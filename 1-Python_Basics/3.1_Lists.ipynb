# Slicing List

```python
numbers = [1, 2, 3, 4, 5]
print(numbers[1:4])  # Output: [2, 3, 4]
```

```python
print(numbers[:5])
```

```python
print(numbers[::2])  # Output: [1, 3, 5]
```

```python
print(numbers[::-1])  # Output: [5, 4, 3, 2, 1]
```

# Iterating Over List

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
```

```python
for fruit in fruits:
    print(fruit)
```

```python
for index, fruit in enumerate(fruits):
    print(f"Index : {index}, Fruits : {fruit}")
```

```python
ls = []
for x in range(10):
    ls.append(x**2)
print(ls)
```

```python
list = [x**2 for x in range(10)]
list
```

# List Comprehension

basic syntax [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

```python
square = [i**2 for i in range(10)]
print(square)
```

```python
lst = []
for i in range(10):
    if i%2 == 0:
        lst.append(i**2)
print(lst)
```

```python
even_numbers = [i for i in range(50) if i%2 != 0]
print(even_numbers)
```

# Nested List Comprehension 

[expression for item1 in iterable1 for item2 in iterable2]

```python
lst1 = [1, 2, 3, 4, 5]
print(lst1)
lst2 = ["a", "b", "c", "d", "e"]
print(lst2)
```

```python
pair = [[i,j] for i in lst1 for j in lst2]
print(pair)
```
