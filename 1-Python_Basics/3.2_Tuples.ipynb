```python
# Pacing in tuples
my_tuple = 1,"Pankaj",3.14,[1, 2, 3],(4, 5)
print(my_tuple)
```

```python
# Unpacking tuples
a,b,c,d,e = my_tuple
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
print(my_tuple)
```

```python
a, *b, c = my_tuple
print(a)
print(b)
print(c)
```

```python
# Nested Tuples
lst = [1, "Pankaj", 3.14,[1, 2000, 3], (4, 5)]
tup = (1, "Pankaj", 3.14,[1, 2, 3], (4, 5))
nested_tuple = (lst, tup)
print(nested_tuple)
```

```python
lst[3][0] = 1000
print(lst)
```

```python
for index, value in enumerate(nested_tuple):
    print(f"Index: {index}, Value: {value}")
```

```python
for item in nested_tuple:
    for sub_item in item:
        print(sub_item, end=" ")
    print()
```
