```python
students = {
    'name': 'John',
    'age': 20,
    'grade': 'A'
}
print(students)
print(students['name'])
```

```python
emp = {
    'name': 'Alice',
    'age': 30,
    'department': 'HR',
    'name': 'Bob'
}
print(emp)
```

```python
# Accessing dictionary values
print(students['age'])
```

```python
# Accessing using get method
print(students.get("grade"))
```

```python
print(students.get("non_existent_key", "Default Value"))
print(students.get("Last_name", "Doe"))
```

```python
# Modifying dictionary values
#Dictionaries are mutable, so we can change their values not keys
students['grade'] = 'A+'    
print(students)
```

```python
students["Address"] = "123 Main St"
print(students)
```

```python
del students['age'] # Deleting a key-value pair
print(students)
```

```python
print(students.keys()) # Get all keys
print(students.values()) # Get all values
print(students.items()) # Get all key-value pairs as tuples
```

```python
students_copy = students
print(students_copy)
print(students)
```

```python
students['name'] = 'Jane'
print(students) 
print(students_copy)  # This will also show the updated name since it's a reference to the same dictionary
```

```python
students_New_copy = students.copy()  # Create a shallow copy of the dictionary
students['name'] = 'Alice'
print(students)
print(students_New_copy)  # This will show the original name 'Jane' since it's a separate copy
```

```python
for key in students.keys():
    print(key)
```

```python
for value in students.values():
    print(value)
```

```python
# Iterating through a dictionary
for key in students:
    print(f"{key}:{students[key]}")
```

```python
# Iterating through a dictionary
for key, value in students.items():
    print(f"{key}:{value}")
```

```python
records = {
    'emp1': {'name': 'Alice', 'age': 30, 'department': 'HR'},
    'emp2': {'name': 'Bob', 'age': 25, 'department': 'IT'},
    'emp3': {'name': 'Charlie', 'age': 35, 'department': 'Finance'}
}

# print(records)
records
```

```python
print(records['emp1']['name'])  # Accessing nested dictionary value
```

```python
# Iterating through a nested dictionary
for emp_id, emp_info in records.items():
    print(f"{emp_id}:{emp_info}")
    for key, value in emp_info.items():
        print(f"  {key}:{value}")
```

```python
# Dictionary comprehension
squared_number = { x: x**2 for x in range(10) if x%2==0}
# squared_number = [ x**2 for x in range(5)]
```

```python
print(squared_number)
```

```python
# Practical example of dictionary comprehension

# use a dictionary to count the frequency of numbers in a list
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

frequency = {num : numbers.count(num) for num in set(numbers)}
print(frequency)
```

```python
freq = {}
for num in numbers:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1
print(freq)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 20, 'c': 30, 'd': 40}
# Merging two dictionaries
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
