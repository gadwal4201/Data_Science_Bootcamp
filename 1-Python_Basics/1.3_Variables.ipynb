# Variables

```python
# Variables in Python
# Variables are used to store data in Python. They are created when you assign a value to them. The variable name can be any valid identifier, and it can hold any type of data, such as numbers, strings, lists, etc.
```

```python
# Declaration and Assignment of Variables
age = 30
height = 5.9
name = "Krish Naik"
is_student = True   

print("Age :", age)
print("Height :", height)
print("Name :", name)
print("Is Student :", is_student)
```

```python
# Naming Conventions in Python
# Variable names should be descriptive
# Variable names should be in lowercase letters and words should be separated by underscores (snake_case)
# Variable names should not start with a number
# Variable names should not be a reserved keyword in Python
# Variable names should not contain special characters except for underscores
```

```python
# Invalid Variable Names
# 1. 2name = "Krish" - Variable names cannot start with a number
# 2. name@ = "Krish" - Variable names cannot contain special characters except
# 3. for = "Krish" - Variable names cannot be a reserved keyword in Python
# 4. name with space = "Krish" - Variable names cannot contain spaces
```

```python
# Python is a dynamically typed language, 
# which means that you do not need to declare the type of a variable when you create it. 
# The type of a variable is determined at runtime based on the value assigned to it. 
# This is known as type inference.
```

```python
# Type Checking and Conversion
# You can check the type of a variable using the type() function.

age = 30
print(type(age))  # Output: <class 'int'>
```

```python
age_str = str(age)
print(age_str, type(age_str))  # Output: 30 <class 'str'>
```

```python
age = "25"
print(type(age))  # Output: <class 'str'>

print(type(int(age)))  # This will raise a ValueError because "Thirty" cannot be converted to an integer
```

```python
# Dynamic Typing in Python

var = 10
print(var, type(var))  # Output: 10 <class 'int'>

var = "Hello"
print(var, type(var))  # Output: Hello <class 'str'>

var = [1, 2, 3]
print(var, type(var))  # Output: [1, 2, 3]

var = {"name": "Krish", "age": 30}
print(var, type(var))  # Output: {'name': 'Krish', 'age
```
